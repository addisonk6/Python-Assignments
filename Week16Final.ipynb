```python
# Expense Tracker

# Use a list to store all expenses
expenses = []

# Main loop:
# show menu (add, totals, exit)
# get user choice
# call appropriate section
# Exit when user chooses
while True:
    print("\n--- Expense Tracker ---")
    print("1. Add Expense")
    print("2. Show Raw Data")
    print("3. Exit")

    choice = input("Selection: ")

    if choice == "1":
        # Add expense:
        # ask for category, vendor, amount, date
        category = input("Enter Category: ")
        vendor = input("Enter Vendor: ")

        # Handle errors for invalid amount input
        try:
            amount = float(input("Enter Amount: "))
        except:
            print("Invalid amount. Try again.")
            continue

        date = input("Enter Date: ")

        # store in dictionary
        expense = {
            "category": category,
            "vendor": vendor,
            "amount": amount,
            "date": date
        }

        # add to list
        expenses.append(expense)
        print("Expense added.")

    elif choice == "2":
        # show stored data (testing core logic)
        print("\nCurrent Expenses:")
        print(expenses)

    elif choice == "3":
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")
```
