```python
# Expense Tracker

# Use a list to store all expenses
expenses = []

# Add expense:
# ask for category, vendor, amount, date
# convert amount to float
# store in dictionary
# add to list
def add_expense():
    category = input("Enter Category: ")
    vendor = input("Enter Vendor: ")

    # Handle errors for invalid amount input
    try:
        amount = float(input("Enter Amount: "))
    except:
        print("Invalid amount. Please enter a number.")
        return

    date = input("Enter Date: ")

    expense = {
        "category": category,
        "vendor": vendor,
        "amount": amount,
        "date": date
    }

    expenses.append(expense)
    print("Expense added successfully.")

# Total logic:
# loop through list
# sum all
# Print totals formatted as $xx.xx
def total_all():
    total = 0
    for exp in expenses:
        total += exp["amount"]
    print(f"Total of all expenses: ${total:.2f}")

# Total logic:
# filter by category
def total_by_category():
    category = input("Enter Category to sum: ")
    total = 0

    for exp in expenses:
        if exp["category"].lower() == category.lower():
            total += exp["amount"]

    print(f"Total for {category}: ${total:.2f}")

# Total logic:
# filter by vendor
def total_by_vendor():
    vendor = input("Enter Vendor to sum: ")
    total = 0

    for exp in expenses:
        if exp["vendor"].lower() == vendor.lower():
            total += exp["amount"]

    print(f"Total for {vendor}: ${total:.2f}")

# Totals menu:
# option for all, category, vendor
def totals_menu():
    print("\n--- Totaling Options ---")
    print("A: All Expenses")
    print("C: Filter by Category")
    print("V: Filter by Vendor")

    choice = input("Choice: ").upper()

    if choice == "A":
        total_all()
    elif choice == "C":
        total_by_category()
    elif choice == "V":
        total_by_vendor()
    else:
        print("Invalid choice.")

# Main loop:
# show menu (add, totals, exit)
# get user choice
# call appropriate section
# Exit when user chooses
while True:
    print("\n--- Expense Tracker ---")
    print("1. Add New Expense")
    print("2. Get Totals")
    print("3. Exit")

    choice = input("Selection: ")

    if choice == "1":
        add_expense()
    elif choice == "2":
        totals_menu()
    elif choice == "3":
        print("Goodbye!")
        break
    else:
        print("Invalid selection.")
```
