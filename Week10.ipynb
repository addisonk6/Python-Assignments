```python
# Name: roster.py


# IMPORT STATMENTS
# (None)



# FUNCTION DEFINITIONS

# add players

def add_players():

# make list called players

  players =["Lebron James", "Stephen Curry", "Kevin Durant", "Victor Wembanyama", "Jason Tatum"]

# ask for player name and if they type "quit", stop loop

  while True:
    name = input("Add a player(or 'quit' to stop): ")

    if name.lower() == "quit": # Fixed: Make 'quit' check case-insensitive
      break

# otherwise add name to list

    players.append(name)

# return the list

  return players

# on_roster(roster, name) and check if name is in roster list
# if it is, return True, if not, return False

def on_roster(roster, name):
  # Fixed: Convert both name and roster names to lower case for case-insensitive comparison
  if name.lower() in [player.lower() for player in roster]:
    return True
  else:
    return False



# PROGRAM LOGIC

# call add_players and store result in roster

roster = add_players() # Fixed: Unindented to be at global scope
print(roster)

# ask user for a name to search

search = input("Who do you want to look for? ")

# call on_roster with roster and name
# if true = print they are on the team, if false = print not on the team

if on_roster(roster, search):
  print("They are on the team.")
else:
  print("They are not on the team.")
```
