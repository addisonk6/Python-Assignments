```python
#IMPORT STATMENTS
#(None)


#FUNCTION DEFINITIONS
def calculate_miles(start_odometer, end_odometer):
    return end_odometer - start_odometer

def calculate_bill(car_ID, days, miles_driven):

  # 1. Get daily rate based on the car ID.
  if car_ID == "B":
      daily_rate = 40
      class_name = "Budget"
  elif car_ID == "S":
      daily_rate = 60
      class_name = "Standard"
  else: # Assuming 'L' for Luxury based on get_valid_car_ID
    daily_rate = 80
    class_name = "Luxury"

  # 4. Get daily rental cost
  rental_cost = daily_rate * days

  # 3. Determine miles charged
  if miles_driven > 100:
        extra_miles = miles_driven - 100
        mileage_cost = extra_miles * 0.35
  else:
        mileage_cost = 0

  # 5. Calculate total bill
  total = rental_cost + mileage_cost

  return total, class_name


#PROGRAM LOGIC

# 1. Ask the user for the car classification ID (B, S, L)
def get_valid_car_ID():
    ID = input("Enter car ID (B = Budget, S = Standard, L = Luxury): ").upper()
    while ID not in ["B", "S", "L"]:
        print("Invalid ID. Please enter B, S, L, or type 'exit' to quit.")
        ID = input("Enter car ID (B = Budget, S = Standard, L = Luxury): ").upper()
        if ID == 'EXIT':
            exit()
    return ID

car_ID = get_valid_car_ID()

# 2. Number of days rented
days = int(input("Enter number of rental days: "))

# 3. Ask for the user start and end odometer reading
start_odometer = int(input("Enter starting odometer reading: "))
end_odometer = int(input("Enter ending odometer reading: "))

# 4. Calculate miles driven
miles_driven = calculate_miles(start_odometer, end_odometer)

# Calculate charged_miles for display purposes, as calculate_bill uses it internally
if miles_driven > 100:
    charged_miles = miles_driven - 100
else:
    charged_miles = 0

# 5. Calculate total bill by calling the function
total_bill, car_class_name = calculate_bill(car_ID, days, miles_driven)

# 7. Display results (Car class name, total miles driven, charged miles, and total bill)
print(f"\n--- Rental Summary ---")
print(f"Car Class: {car_class_name}")
print(f"Total Miles Driven: {miles_driven}")
print(f"Charged Miles: {charged_miles}")
print(f"Total Bill: ${total_bill:.2f}")
```
