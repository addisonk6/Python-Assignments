```python
# NAME: library.py
# IMPORT STATEMENTS

# CLASS DEFINITIONS
class Book:
    # __init__(title, author)
    def __init__(self, title, author):

    # - save title
      self.title = title

    # - save author
      self.author = author

    # - set is_available to True
      self.is_available = True


    # __str__()
    def __str__(self):

    # - set status if
      if self.is_available:

    # - return "title by author - status"
        status = "Available"

    # checkout()
      else:
        status = "Checked Out"

    # - set is_available to False

    # return_book()
      return f"{self.title} by {self.author} - {status}"

        # default status
    def checkout(self):
      self.is_available = False

    def return_book(self):
      self.is_available = True

# PROGRAM LOGIC
# - create 2 books
book1 = Book("Think Python", "Allen Downey")
book2 = Book("The Hobbit", "J.R.R. Tolkien")

# - print both
print("--- Library Status ---")
print(book1)
print(book2)

# - check out 1 book
print("\nChecking out 'The Hobbit'...\n")
book2.checkout()

# - print both again
print("--- Updated Status ---")
print(book1)
print(book2)
```
