```python
# NAME: journal.py

# IMPORTS

# none needed


# FUNCTIONS

# write_entry
# take filename + entry

def write_entry(filename, entry):

# open file in 'a'
# write entry + \n

    with open(filename, 'a') as file:
        file.write(entry + "\n") # new line

# read_journal
# take filename
# open file in 'r'
# print contents


# PROGRAM LOGIC

# set filename

filename = "journal.txt"

# loop:
#   ask W / R / Q
#   if W = get input, save
#   if R = read file
#   if Q = quit
#   else = error message

while True:
    choice = input("(W)rite, (R)ead, or (Q)uit? ").upper()

    if choice == 'W':
        entry = input("Enter your entry: ")
        write_entry(filename, entry)
        print("Entry saved.")

    elif choice == 'R':
        print("Not added yet.")

    elif choice == 'Q':
        print("Goodbye!")
        break

    else:
        print("Invalid option.")
```
