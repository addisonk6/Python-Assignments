```python
# ==========================================================
# Rental Car Pricing System
# ==========================================================


def calculate_bill(car_code, days_rented, start_odometer, end_odometer):

    # Determine daily rate and class name
    if car_code == "B":
        daily_rate = 40
        class_name = "Budget"

    elif car_code == "S":
        daily_rate = 60
        class_name = "Standard"

    else:
        daily_rate = 80
        class_name = "Luxury"


    # Calculate miles driven
    miles_driven = end_odometer - start_odometer


    # Determine charged miles
    if miles_driven > 100:
        charged_miles = miles_driven - 100
    else:
        charged_miles = 0


    # Mileage cost
    mileage_cost = charged_miles * 0.35


    # Rental cost
    rental_cost = days_rented * daily_rate


    # Total cost
    total_cost = rental_cost + mileage_cost


    return total_cost, miles_driven, charged_miles, class_name


# ----------------------------------------------------------
# MAIN PROGRAM
# ----------------------------------------------------------

# Get valid car code
while True:
    car_code = input("Enter car class (B = Budget, S = Standard, L = Luxury): ").upper()

    if car_code in ["B", "S", "L"]:
        break
    else:
        print("Invalid code. Please enter B, S, or L.")


# Get days rented
days_rented = int(input("Enter number of rental days: "))


# Get odometer readings
start_odometer = int(input("Enter starting odometer reading: "))
end_odometer = int(input("Enter ending odometer reading: "))


# Call function
total_cost, miles_driven, charged_miles, class_name = calculate_bill(
    car_code, days_rented, start_odometer, end_odometer
)

# Output results
print("\n----- Rental Summary -----")
print("Car class:", class_name)
print("Miles driven:", miles_driven)
print("Charged miles:", charged_miles)
print("Total bill: $", round(total_cost, 2))
```
